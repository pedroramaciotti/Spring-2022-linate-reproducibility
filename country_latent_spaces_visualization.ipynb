```python
# visualization of three dimensions

xdim = 'latent_dimension_0'
ydim = 'latent_dimension_1'
zdim = 'latent_dimension_2'

ren = {xdim: 'x', ydim: 'y', zdim: 'z'}

f_df = followers_df.rename(columns = ren).drop_duplicates(subset = ['x', 'y', 'z'])
p_df = parliamentarians_df.rename(columns = ren).drop_duplicates(subset = ['x', 'y', 'z'])
df = pd.concat([f_df, p_df], axis = 0)
df = df.drop_duplicates(subset = ['x', 'y', 'z'])
df = df[~df.index.duplicated()]
```

```python
# plot first three latent dimensions in 3D

threedimensional_figsize = (6, 5.2)
threedimensional_label_fs = 12
threedimensional_title_fs = 18

fig_df = f_df.sample(n = 10000)  # use a sample of the dataset
    
fig = plt.figure(figsize = threedimensional_figsize)

ax = fig.add_subplot(1, 1, 1, projection = '3d')

ax.scatter(fig_df['x'], fig_df['y'], fig_df['z'], marker = 's', c = fig_df['z'], alpha = 0.5, s = 3)

ax.set_xlabel(r'$\delta^*_1$', fontsize = threedimensional_label_fs)
ax.set_ylabel(r'$\delta^*_2$', fontsize = threedimensional_label_fs)
ax.set_zlabel(r'$\delta^*_3$', fontsize = threedimensional_label_fs)

ax.set_title(country +' in latent space',fontsize=threedimensional_title_fs)

plt.tight_layout()

plt.show()

#plt.savefig('Figures/LatentSpace/%s_latent_3d.png'%country,dpi=300)
#plt.clf()
#plt.close()   
```
